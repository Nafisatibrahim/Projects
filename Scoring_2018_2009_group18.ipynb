```python
# Import Python packages
import pandas as pd

# Load datasets for Scoring statistics 2009-2018
df1 = pd.read_csv('2018 Big-Ten Scoring.csv')
df2 = pd.read_csv('2017 Big-Ten Scoring.csv')
df3 = pd.read_csv('2016 Big-Ten Scoring.csv')
df4 = pd.read_csv('2015 Big-Ten Scoring.csv')
df5 = pd.read_csv('2014 Big-Ten Scoring.csv')
df6 = pd.read_csv('2013 Big-Ten Scoring.csv')
df7 = pd.read_csv('2012 Big-Ten Scoring.csv')
#df8 = pd.read_csv('2011 Big-Ten Scoring.csv')
#df9 = pd.read_csv('2010 Big-Ten Scoring.csv')
#df10 = pd.read_csv('2009 Big-Ten Scoring.csv')
```

<a name="section2"></a>
### 2. Aggregating Team Score


```python
# Aggregate scoring data by team
def team_score(df):
    aggregated_df = df.groupby('Team')['Pts'].sum().reset_index()
    return aggregated_df

# Apply the aggregation function to each DataFrame
df1_agg = team_score(df1)
df2_agg = team_score(df2)
df3_agg = team_score(df3)
df4_agg = team_score(df4)
df5_agg = team_score(df5)
df6_agg = team_score(df6)
df7_agg = team_score(df7)
#df8_agg = team_score(df8)
#df9_agg = team_score(df9)
#df10_agg = team_score(df10)

```

```python
# Display team score for year 2018
df1_agg
```

```python
# Concatenate all aggregated DataFrames
all_teams_df = pd.concat([df1_agg, df2_agg, df3_agg, df4_agg, df5_agg, df6_agg, df7_agg])

# Select unique team names
unique_teams = all_teams_df['Team'].unique()

# List all unique teams
list(unique_teams)
```

```python
# Add a 'Year' column to each aggregated DataFrame
df1_agg['Year'] = '2018'
df2_agg['Year'] = '2017'
df3_agg['Year'] = '2016'
df4_agg['Year'] = '2015'
df5_agg['Year'] = '2014'
df6_agg['Year'] = '2013'
df7_agg['Year'] = '2012'
#df8_agg['Year'] = '2011'
#df9_agg['Year'] = '2010'
#df10_agg['Year'] = '2009'


# Concatenate all aggregated DataFrames into a single table
full_table = pd.concat([df1_agg, df2_agg, df3_agg, df4_agg, df5_agg, df6_agg, df7_agg])

# Rearrange the columns to match the requested format: Year, Team, Scoring
final_table = full_table[['Year', 'Team', 'Pts']].rename(columns={'Pts': 'Scoring'})

# Display the final table
print(final_table)
```

###3. Downloading Final Table


```python
from google.colab import files

# Save the DataFrame to a CSV file
final_table.to_csv('final_scoring_table_2018-2009.csv', index=False)

# Download the final table
files.download('final_scoring_table_2018-2009.csv')
```
